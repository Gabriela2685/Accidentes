```python
import pandas as pd
```

```python
vehiculos = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vTFDyV7w_ZA_0Wa9TMT2iISK2cExw4leMBHXaN6CH11Ct_j_L5vOzxoDq25SnBgBw/pub?output=csv')
gravedad = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vT1l2Zghki6yHvfnvkxEeJTyfMQQ-O2UQX_x6-NJJg9edfGZbWgbGXrfxwn5wVMhw/pub?output=csv')
```

```python
# Realiza un merge en la columna de folio para obtener los registros comunes
folios_comunes = pd.merge(vehiculos, gravedad, on="folio")
```

```python
# Guardar el resultado en un archivo de Excel
folios_comunes.to_csv("folios_comunes.csv", index=False)
```
